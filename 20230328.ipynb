```python
import pandas as pd

def data_cleaning(filename):
    book_info = ['도서명', '저자', 'ISBN', '권']
    ns_df = pd.read_csv(filename, low_memory=False)
    ns_book = ns_df.dropna(axis=1, how='all')
    count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
    loac_count = count_df.groupby(by=book_info, dropna=False).sum()
    dup_rows = ns_book.duplicated(subset=book_info)
    unique_rows = ~dup_rows
    ns_book3 = ns_book[unique_rows].copy()
    ns_book3.set_index(book_info, inplace=True)
    ns_book3.update(loac_count)
    ns_book4 = ns_book3.reset_index()
    ns_book4 = ns_book4[ns_book.columns]
    return ns_book4
```

```python
new_ns_book4 = data_cleaning('/Users/yunseokgyu/study/school-class/ml-vocational-class/ns_202104.csv')
ns_book4.equals(new_ns_book4)
```
